```python
# count and percentage of Orders purchased in the month of Jan 2018 with 5 Review score

sql='''

with data as
(select distinct o.customer_id, o.order_id, r.review_id, r.review_score, o.order_delivered_customer_date
from olist_orders o
join olist_order_reviews r on o.order_id = r.order_id
where o.order_delivered_customer_date between '2018-01-01 00:00:00' and '2018-01-31 23:59:59'),

data2 as
(select (select count(1) from data where review_score = 5) as order_with_5_review, count(1) as total_orders 
from data)

select *,
(cast(order_with_5_review as float)/total_orders)*100 as percentage
from data2

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Customer purchase trend YoY

sql='''

with data as
(select distinct customer_id, order_id, order_delivered_customer_date, order_status,
strftime('%Y',order_delivered_customer_date) as year,
strftime('%m',order_delivered_customer_date) as month
from olist_orders
where order_status= 'delivered' and order_delivered_customer_date is not null
order by order_delivered_customer_date)

select month, case 
when month='01' then 'Jan' 
when month='02' then 'Feb'
when month='03' then 'Mar'
when month='04' then 'Apr'
when month='05' then 'May'
when month='06' then 'Jun'
when month='07' then 'Jul'
when month='08' then 'Aug'
when month='09' then 'Sep'
when month='10' then 'Oct'
when month='11' then 'Nov'
when month='12' then 'Dec'
else 0 end as month_name,
sum(case when year= '2016' then 1 else 0 end) as Year_2016,
sum(case when year= '2017' then 1 else 0 end) as Year_2017,
sum(case when year= '2018' then 1 else 0 end) as Year_2018

from data
group by month
order by month

''';

df_sql = pd.read_sql_query(sql,con = engine)
df_sql
```

```python
# Average order values of customers

sql='''

with delivery_details as

(select distinct c.customer_unique_id, c.customer_id, o.order_id, c.customer_city, c.customer_state, o.order_status, o.order_delivered_customer_date
from olist_customers c
join olist_orders o on c.customer_id = o.customer_id),

payment_details as
(select distinct o.customer_id, o.order_id, o.order_status, o.order_delivered_customer_date, p.payment_value
from olist_orders o
join olist_order_payments p on o.order_id=p.order_id),

raw_data as
(select distinct delivery_details.customer_unique_id, delivery_details.customer_id, delivery_details.order_id, delivery_details.customer_city, delivery_details.customer_state, delivery_details.order_status, delivery_details.order_delivered_customer_date, payment_details.payment_value
from delivery_details
join payment_details on
delivery_details.customer_id = payment_details.customer_id
and delivery_details.order_id = payment_details.order_id 
and delivery_details.order_status = payment_details.order_status
and delivery_details.order_delivered_customer_date = payment_details.order_delivered_customer_date),

order_values as
(select customer_unique_id, count(order_id) as Total_Orders_By_Customers, avg(payment_value) as Total_Payment_By_Customers
from raw_data
where order_status = 'delivered'
group by 1
order by 1)

select v.customer_unique_id, v.Total_Orders_By_Customers, v.Total_Payment_By_Customers, c.customer_city, c.customer_state
from order_values v
join olist_customers c on c.customer_unique_id = v.customer_unique_id

''';
    
df_sql = pd.read_sql_query(sql,con = engine)
df_sql
```

```python
# Top 10 Cities with highest revenue


sql = '''

with delivery_details as

(select distinct c.customer_unique_id, c.customer_id, o.order_id, c.customer_city, c.customer_state, o.order_status, o.order_delivered_customer_date
from olist_customers c
join olist_orders o on c.customer_id = o.customer_id),

payment_details as
(select distinct o.customer_id, o.order_id, o.order_status, o.order_delivered_customer_date, p.payment_value
from olist_orders o
join olist_order_payments p on o.order_id=p.order_id),

raw_data as
(select distinct delivery_details.customer_unique_id, delivery_details.customer_id, delivery_details.order_id, delivery_details.customer_city, delivery_details.customer_state, delivery_details.order_status, delivery_details.order_delivered_customer_date, payment_details.payment_value
from delivery_details
join payment_details on
delivery_details.customer_id = payment_details.customer_id
and delivery_details.order_id = payment_details.order_id 
and delivery_details.order_status = payment_details.order_status
and delivery_details.order_delivered_customer_date = payment_details.order_delivered_customer_date)

select customer_city, customer_state, count(distinct order_id) as Total_Orders_By_Customers, sum(payment_value) as Total_Payment_By_Customers
from raw_data
where order_status = 'delivered'
group by 1,2
order by 4 desc
limit 10

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Top Successfull sellers in terms of no of goods sold, total revenue, customer_count, selleer with highest 5 star ratings

sql='''

with raw as
(select distinct c.customer_id, c.customer_unique_id, s.seller_id, o.order_id, o.order_status, i.order_item_id, i.product_id, i.price, i.freight_value, i.price - i.freight_value as profit, r.review_score, s.seller_city, s.seller_state, c.customer_city, c.customer_state, o.order_delivered_customer_date
from olist_orders o
join olist_customers c on o.customer_id = c.customer_id
join olist_order_items i on o.order_id = i.order_id
join olist_sellers s on i.seller_id = s.seller_id
join olist_order_reviews r on o.order_id = r.order_id
where order_status = 'delivered')

select 'Seller with most of unique customers :-' as Description, seller_id, count(distinct customer_unique_id) as Value
from raw

union all

select 'Seller with highest Profit :-' as Description, seller_id, sum(profit) as Value
from raw

union all

select 'Seller with most unique orders :-' as Description, seller_id, count(distinct order_id) as Value
from raw

union all

select 'Seller with most 5 star ratings :-' as Description, a.seller_id, count(a.review_score) as Value
from
(select distinct c.customer_id, c.customer_unique_id, s.seller_id, o.order_id, o.order_status, i.order_item_id, i.product_id, i.price, i.freight_value, i.price - i.freight_value as profit, r.review_score, s.seller_city, s.seller_state, c.customer_city, c.customer_state, o.order_delivered_customer_date
from olist_orders o
join olist_customers c on o.customer_id = c.customer_id
join olist_order_items i on o.order_id = i.order_id
join olist_sellers s on i.seller_id = s.seller_id
join olist_order_reviews r on o.order_id = r.order_id
where order_status = 'delivered'
and review_score >= 5) a

''';
    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Delivery Success Rate across state

sql='''

with data as
(select c.customer_state,
strftime('%Y', o.order_delivered_customer_date) as Year,
count(distinct c.customer_city) as Total_Cities_Delivered,
count(c.customer_unique_id) as Total_Customers,
sum(case when o.order_status = 'delivered' then 1 else 0 end) as Delivered,
sum(case when o.order_status <> 'delivered' then 1 else 0 end) as Not_Delivered
from olist_customers c
join olist_orders o on o.customer_id = c.customer_id
where Year is not null
group by 1, 2
order by 1, 2)

select *,
round((cast(Delivered as float)/Total_Customers)*100, 2) as Delivery_Success_Rate,
round((cast(Not_Delivered as float)/Total_Customers)*100, 2) as Delivery_Failure_Rate
from data

''';
    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Preferred form of payment for differenct categories

sql='''

with raw as
(select pre.product_category_name_english as product_category_name, p.order_id, p.payment_type
from olist_order_payments p
join olist_orders o on o.order_id = p.order_id
join olist_order_items i on i.order_id = o.order_id
join olist_products pr on i.product_id = pr.product_id
join product_category_name_translation pre on pre.product_category_name = pr.product_category_name
group by 1,2,3),

result as
(select product_category_name,
case when payment_type='credit_card' then count(order_id) else 0 end as Credit_Card_Payment,
case when payment_type='boleto' then count(order_id) else 0 end as Boleto_Payment,
case when payment_type='voucher' then count(order_id) else 0 end as Voucher_Payment,
case when payment_type='debit_card' then count(order_id) else 0 end as Debit_Card_Payment,
case when payment_type='not_defined' then count(order_id) else 0 end as Not_defined_Payment
from raw
group by 1)

select 
product_category_name as Category_Name,
case max(Credit_Card_Payment, Boleto_Payment, Voucher_Payment, Debit_Card_Payment, Not_defined_Payment) 

when Credit_Card_Payment then 'Credit_Card_Payment'
when Boleto_Payment then 'Boleto_Payment'
when Voucher_Payment then 'Voucher_Payment'
when Debit_Card_Payment then 'Debit_Card_Payment'
when Not_defined_Payment then 'Not_defined_Payment'
end as Preferred_option,

case max(Credit_Card_Payment, Boleto_Payment, Voucher_Payment, Debit_Card_Payment, Not_defined_Payment) 

when Credit_Card_Payment then max(Credit_Card_Payment)
when Boleto_Payment then max(Boleto_Payment)
when Voucher_Payment then max(Voucher_Payment)
when Debit_Card_Payment then max(Debit_Card_Payment)
when Not_defined_Payment then max(Not_defined_Payment)
end as Total_Payments
from result
group by 1



''';
    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```
