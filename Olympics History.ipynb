```python
import pandas as pd
import numpy as np
import sqlite3 as sql
from sqlalchemy import create_engine
```

```python
athletes = pd.read_csv('athlete_events.csv')
noc = pd.read_csv('noc_regions.csv')
```

```python
olympics_df = pd.merge(athletes, noc, on = 'NOC')
olympics_df = olympics_df.rename(columns = {'region' : 'Region', 'notes' : 'Notes'})
olympics_df
```

```python
engine = create_engine('sqlite://', echo = False)

# export the dataframe as a table to the sqlite engine
olympics_df.to_sql('olympics', con = engine)
```

```python
# select some entries from table to test the code

sql='''

select *
from olympics
limit 10

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# total olympic games held

sql='''

select count(distinct games) as total_olympic_games
from olympics

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# all olympic games held so far

sql='''

select distinct games, year, season, city
from olympics
order by year

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# total number of nations that participated in each games

sql='''

with data as
(select games, region
from olympics
group by games, region)

select games, count(1) as total_countries
from data
group by games
order by games

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Which year saw the highest and lowest no of countries participating 

sql='''

with data as
(select games, region
from olympics
group by games, region),

data2 as
(select games, count(1) as total_countries
from data
group by games
order by games)

select distinct
first_value(games) over (order by total_countries)||' - '||first_value(total_countries) over (order by total_countries) as lowest_countries,
first_value(games) over (order by total_countries desc)||' - '||first_value(total_countries) over (order by total_countries desc) as highest_countries
from data2

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# which nation participated in all the olympic games

sql = '''

with data as
(select region, games
from olympics
group by region, games)

select region as country, count(1) as total_participated_games
from data
group by region
having count(1) = (select count(distinct games) from olympics)


''';
  
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# sport which was played in all summer olympics

sql = '''

with data as
(select sport, games
from olympics
where season = 'Summer'
group by sport, games)

select sport, count(1) as no_of_games
from data
group by sport
having count(1) = (select count(distinct games) from olympics where season = 'Summer') 

''';
  
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# sport which was played only once in olympics

sql = '''

with data as
(select sport, games
from olympics
group by sport, games),

data2 as
(select sport, count(1) as no_of_games
from data
group by sport
having count(1) = 1)

select data2.sport, no_of_games, games
from data
join data2 on data.sport = data2.sport

''';
  
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# total number of sports played in each olympic games

sql = '''

with data as
(select games, sport
from olympics
group by games, sport)

select games, count(1) as no_of_sports
from data
group by games
order by 2 desc

''';
  
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# oldest athlete to win a gold medal

sql = '''

with data as
(select *
from olympics
where medal = 'Gold')

select *
from data
where age = (select max(age) from data)

''';
  
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# ratio of female to male athletes participated in all olympic games

sql = '''

with female as
(select count(1) as female_athletes from olympics where sex = 'F'),

male as
(select count(1) as male_athletes from olympics where sex = 'M')

select '1 : '||round((cast(male_athletes as float)/female_athletes), 2) as ratio
from female
join male

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# top 5 athletes who have won most gold medals

sql = '''

with data as
(select name, team, count(1) as total_gold_medals
from olympics
where medal = 'Gold'
group by name, team),

data2 as
(select *,
dense_rank() over (order by total_gold_medals desc) as rnk
from data
)

select name, team, total_gold_medals
from data2
where rnk <= 5

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# top 5 athletes who have won most medals

sql = '''

with data as
(select name, team, count(1) as total_gold_medals
from olympics
where medal in ('Gold', 'Silver', 'Bronze')
group by name, team),

data2 as
(select *,
dense_rank() over (order by total_gold_medals desc) as rnk
from data
)

select name, team, total_gold_medals
from data2
where rnk <= 5

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# top 5 most successful countries

sql = '''

with data as
(select region, count(1) as total_medals
from olympics
where medal is not null
group by region),

data2 as 
(select dense_rank() over (order by total_medals desc) as rank, *
from data)

select * 
from data2
where rank <= 5

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# total gold, silver, bronze won by each country

sql = '''

select region as country,
sum(case when medal = 'Gold' then 1 else 0 end) as gold,
sum(case when medal = 'Silver' then 1 else 0 end) as silver,
sum(case when medal = 'Bronze' then 1 else 0 end) as bronze
from olympics
group by region 
order by 2 desc, 3 desc, 4 desc

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# total gold, silver, bronze won by each country in each olympic games

sql = '''

select games, region as country,
sum(case when medal = 'Gold' then 1 else 0 end) as gold,
sum(case when medal = 'Silver' then 1 else 0 end) as silver,
sum(case when medal = 'Bronze' then 1 else 0 end) as bronze
from olympics
group by games, region
order by 1, 3 desc, 4 desc, 5 desc

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# country that won the most gold, most silver and most bronze medals in each olympic games

sql = '''

with data as
(select games, region as country,
sum(case when medal = 'Gold' then 1 else 0 end) as gold,
sum(case when medal = 'Silver' then 1 else 0 end) as silver,
sum(case when medal = 'Bronze' then 1 else 0 end) as bronze
from olympics
group by games, region
order by 1, 3 desc, 4 desc, 5 desc)

select distinct games,
first_value(country) over(partition by games order by gold desc) || ' - ' || first_value(gold) over(partition by games order by gold desc) as max_gold,
first_value(country) over(partition by games order by silver desc) || ' - ' || first_value(silver) over(partition by games order by silver desc) as max_silver,
first_value(country) over(partition by games order by bronze desc) || ' - ' || first_value(bronze) over(partition by games order by bronze desc) as max_bronze
from data
order by games

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# country that won the most gold, most silver, most bronze and overall most medals in each olympic games

sql = '''

with data as
(select games, region as country,
sum(case when medal = 'Gold' then 1 else 0 end) as gold,
sum(case when medal = 'Silver' then 1 else 0 end) as silver,
sum(case when medal = 'Bronze' then 1 else 0 end) as bronze
from olympics
group by games, region
order by 1, 3 desc, 4 desc, 5 desc),

data2 as
(select *, gold + silver + bronze as total_medal
from data)

select distinct games,
first_value(country) over(partition by games order by gold desc) || ' - ' || first_value(gold) over(partition by games order by gold desc) as max_gold,
first_value(country) over(partition by games order by silver desc) || ' - ' || first_value(silver) over(partition by games order by silver desc) as max_silver,
first_value(country) over(partition by games order by bronze desc) || ' - ' || first_value(bronze) over(partition by games order by bronze desc) as max_bronze,
first_value(country) over(partition by games order by total_medal desc) || ' - ' || first_value(bronze) over(partition by games order by total_medal desc) as max_bronze
from data2
order by games

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# which country never won gold but have won silver/gold

sql = '''

with data as
(select region as country,
sum(case when medal = 'Gold' then 1 else 0 end) as gold,
sum(case when medal = 'Silver' then 1 else 0 end) as silver,
sum(case when medal = 'Bronze' then 1 else 0 end) as bronze
from olympics
group by region 
order by 2 desc, 3 desc, 4 desc)

select * 
from data
where gold = 0 and (silver <> 0 or bronze <> 0)


''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# which country never won any medal

sql = '''

with data as
(select region as country,
sum(case when medal = 'Gold' then 1 else 0 end) as gold,
sum(case when medal = 'Silver' then 1 else 0 end) as silver,
sum(case when medal = 'Bronze' then 1 else 0 end) as bronze
from olympics
group by region 
order by 2 desc, 3 desc, 4 desc)

select * 
from data
where gold = 0 and silver = 0 and bronze = 0 
and country is not null


''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```
