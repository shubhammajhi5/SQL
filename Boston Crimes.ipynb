```python
# query to show count of auto theft and towed by phase of day vs month 

sql='''

with data as
(Select *,
substr(OCCURRED_ON_DATE, 12) as time,
case 
when substr(OCCURRED_ON_DATE, 12) between '06:00:00' and '11:00:00' then 'Morning'
when substr(OCCURRED_ON_DATE, 12) between '11:01:00' and '17:00:00' then 'Noon'
when substr(OCCURRED_ON_DATE, 12) between '17:01:00' and '20:00:00' then 'Evening'
when substr(OCCURRED_ON_DATE, 12) between '20:01:00' and '23:59:00' then 'Night'
when substr(OCCURRED_ON_DATE, 12) between '00:00:00' and '05:59:00' then 'Night'
else 'Not Available'
end as phase_of_day
from crime)

select phase_of_day,
sum(case when month = 1 then 1 else 0 end) as Jan,
sum(case when month = 2 then 1 else 0 end) as Feb,
sum(case when month = 3 then 1 else 0 end) as Mar,
sum(case when month = 4 then 1 else 0 end) as Apr,
sum(case when month = 5 then 1 else 0 end) as May,
sum(case when month = 6 then 1 else 0 end) as Jun,
sum(case when month = 7 then 1 else 0 end) as Jul,
sum(case when month = 8 then 1 else 0 end) as Aug,
sum(case when month = 9 then 1 else 0 end) as Sep,
sum(case when month = 10 then 1 else 0 end) as Oct,
sum(case when month = 11 then 1 else 0 end) as Nov,
sum(case when month = 12 then 1 else 0 end) as Dec
from data
where OFFENSE_CODE_GROUP in ('Auto Theft', 'Towed')
group by phase_of_day



''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
## script to get full offence name per district which has maximum occurence in respective district

sql='''

with data as
(Select district, offense_code, count(1) as no_of_crime,
rank() over (partition by district order by count(1) desc) as rank
from crime
where district is not null
group by district, offense_code)

select distinct district, offense_code, name, no_of_crime
from data d
join offense_codes o on o.code = d.offense_code
where rank = 1
order by district

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
## add a column which contains date of last incidents that happened in respective district

sql='''

Select *,
lag(OCCURRED_ON_DATE) over (partition by DISTRICT order by OCCURRED_ON_DATE) as last_incident_on
from crime
where district is not null
order by district, OCCURRED_ON_DATE


''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(50)
```
