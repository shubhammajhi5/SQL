```python
import pandas as pd
from pandasql import sqldf
```

```python
olympics = pd.read_csv('athlete_events.csv')
```

```python
olympics.head()
```

```python
# select some entries from table to test the code

query = '''select *
           from olympics
           limit 10'''

df = sqldf(query)
df
```

```python
# total olympic games held

query = '''select count(distinct games) as total_olympic_games
           from olympics'''

df = sqldf(query)
df
```

```python
# all olympic games held so far

query = '''select distinct games, year, season, city
           from olympics
           order by year'''

df = sqldf(query)
df
```

```python
# total number of nations that participated in each games

query = '''with data as
           (select games, NOC
           from olympics
           group by games, NOC)

           select games, count(1) as total_countries
           from data
           group by games
           order by games'''

df = sqldf(query)
df
```

```python
# Which year saw the highest and lowest no of countries participating 

query = '''with data as
           (select games, NOC
           from olympics
           group by games, NOC),

           data2 as
           (select games, count(1) as total_countries
           from data
           group by games
           order by games)

           select distinct
           first_value(games) over (order by total_countries)||' - '||first_value(total_countries) over (order by total_countries) as lowest_countries,
           first_value(games) over (order by total_countries desc)||' - '||first_value(total_countries) over (order by total_countries desc) as highest_countries
           from data2'''

df = sqldf(query)
df
```

```python
# which nation participated in all the olympic games

query = '''with data as
           (select NOC, games
           from olympics
           group by NOC, games)

           select NOC as country, count(1) as total_participated_games
           from data
           group by NOC
           having count(1) = (select count(distinct games) from olympics)'''

df = sqldf(query)
df
```

```python
# sport which was played in all summer olympics

query = '''with data as
           (select sport, games
           from olympics
           where season = 'Summer'
           group by sport, games)

           select sport, count(1) as no_of_games
           from data
           group by sport
           having count(1) = (select count(distinct games) from olympics where season = 'Summer') '''

df = sqldf(query)
df
```

```python
# sport which was played only once in olympics

query = '''with data as
           (select sport, games
           from olympics
           group by sport, games),

           data2 as
           (select sport, count(1) as no_of_games
           from data
           group by sport
           having count(1) = 1)

           select data2.sport, no_of_games, games
           from data
           join data2 on data.sport = data2.sport'''

df = sqldf(query)
df
```
