```python
import pandas as pd
import numpy as np
import sqlite3 as sql
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('Salary Dataset.csv')
df
```

```python
df = df.rename(columns = {'Company Name': 'Company_Name', 
                          'Job Title': 'Job_Title',
                         'Salaries Reported': 'Salaries_Reported'
                         })
df
```

```python
df.info()
```

```python
engine = create_engine('sqlite://', echo = False)

# export the dataframe as a table to the sqlite engine
df.to_sql('Salary_Dataset', con = engine)
```

```python
# select some entries from table to test the code

sql='''

select *
from Salary_Dataset
limit 10

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# total no of companies

sql='''

select count(distinct company_name) as no_of_companies
from Salary_Dataset

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# ibm data

sql='''

select *
from Salary_Dataset
where company_name = 'IBM'

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# different job titles available

sql='''

select distinct job_title
from Salary_Dataset

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# no of companies in each city

sql = '''

select location, count(distinct company_name) as no_of_companies
from Salary_Dataset
group by location
order by 2 desc

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Count of salariesReported at a job description level

sql='''

select job_title, count(salaries_reported) as count_of_reports
from Salary_Dataset
group by job_title
order by 2 desc

''';

    
df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# company size identification

sql='''

with data as
(select company_name, count(distinct location) as no_of_branches
from Salary_Dataset
where company_name is not null
group by company_name)

select company_name, no_of_branches,
case
when no_of_branches > 3 then 'Big'
when no_of_branches in (2, 3) then 'Medium'
when no_of_branches = 1 then 'Small'
else 'No Branch'
end as size_of_company
from data
order by 2 desc

''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Count of branches at a job description level for all the cities

sql='''

with data as
(select 
company_name,
location,
count(distinct location) as count_of_branches
from Salary_Dataset
where company_name is not null
group by 1,2)

select location,
sum(case when company_name='Mu Sigma' then 1 else 0 end) as Mu_Sigma,
sum(case when company_name='IBM' then 1 else 0 end) as IBM,
sum(case when company_name='Tata Consultancy Services' then 1 else 0 end) as Tata_Consultancy_Services,
sum(case when company_name='Impact Analytics' then 1 else 0 end) as Impact_Analytics,
sum(case when company_name='Accenture' then 1 else 0 end) as Accenture,
sum(case when company_name='Infosys' then 1 else 0 end) as Infosys,
sum(case when company_name='Capgemini' then 1 else 0 end) as Capgemini,
sum(case when company_name='Cognizant Technology Solutions' then 1 else 0 end) as Cognizant_Technology_Solutions
from data
group by location
''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```
